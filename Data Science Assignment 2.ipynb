```python
# We take a subset of the variables which are important for model building
data=df.drop(['Current City','Other skills', 'Institute', 'Degree','Stream','Performance_PG',
              'Performance_UG','Performance_12','Performance_10'],axis=1)
```

```python
data['grandtotal']= 40
data.head()
```

```python
data[(data['grandtotal'] >= 40)]
```

```python
#making a result column in order to show shortlisted/not shortlisted candidates
data.criterion = 'NaN'
for i, row in data.iterrows() :
    if (row['grandtotal'] >=40):
        data.at[i,'criterion']="Shortlisted"
    else :
        data.at[i,'criterion']="Not Shortlisted"
```

```python
inputs = data.drop(['criterion'],axis=1)
target=data['criterion']
output=data.criterion
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier(criterion='entropy')
```

```python
x_train,x_test,y_train,y_test=train_test_split(inputs,target)
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```
